```python
#now i'll try to do some transformations on the data to see if we can improve the model
X.columns
```

```python
X['LH/FSH Ratio'] = X['LH(mIU/mL)'] / X['FSH(mIU/mL)']

X['Systolic/Diastolic Ratio'] = X['BP _Systolic (mmHg)'] / X['BP _Diastolic (mmHg)']

X['FollicleSym'] = (X['Follicle No. (L)'] - X['Follicle No. (R)']).abs()

X['Avg F Size'] = X['Avg. F size (L) (mm)'] / X['Avg. F size (R) (mm)']

X['Beta-HCG Difference'] = (X['II    beta-HCG(mIU/mL)_x'] - X['II    beta-HCG(mIU/mL)_y']).abs()

X['Weight-Age Ratio'] = X['Weight (Kg)'] / X['Age (yrs)']

X['Height-Age Ratio'] = X['Height(Cm)'] / X['Age (yrs)']

X['Total Hormonal Symptoms Count'] = X[['Pregnant(Y/N)', 'Weight gain(Y/N)', 'hair growth(Y/N)',
                                        'Skin darkening (Y/N)', 'Hair loss(Y/N)', 'Pimples(Y/N)',
                                        'Fast food (Y/N)', 'Reg.Exercise(Y/N)']].replace({'Y':1,'N':0}).sum(axis=1)
```
